```python
import os
import shutil
import random

# المسار الخاص بك
original_data = '/content/drive/MyDrive/DentalDataSet'
base_dir = '/content/drive/MyDrive/Dental_Final_Split'

classes = os.listdir(original_data)

for split in ['train', 'val', 'test']:
    for cls in classes:
        os.makedirs(os.path.join(base_dir, split, cls), exist_ok=True)

for cls in classes:
    src_path = os.path.join(original_data, cls)
    images = [f for f in os.listdir(src_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    random.shuffle(images)

    train_idx = int(len(images) * 0.8)
    val_idx = int(len(images) * 0.9)

    train_imgs = images[:train_idx]
    val_imgs = images[train_idx:val_idx]
    test_imgs = images[val_idx:]

    for img in train_imgs: shutil.copy(os.path.join(src_path, img), os.path.join(base_dir, 'train', cls, img))
    for img in val_imgs: shutil.copy(os.path.join(src_path, img), os.path.join(base_dir, 'val', cls, img))
    for img in test_imgs: shutil.copy(os.path.join(src_path, img), os.path.join(base_dir, 'test', cls, img))

print("✅ تم التقسيم بنجاح. الداتا الآن موزعة في: Dental_Final_Split")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import numpy as np

# إعدادات الـ Augmentation
datagen = ImageDataGenerator(
    rotation_range=30, width_shift_range=0.2, height_shift_range=0.2,
    zoom_range=0.2, horizontal_flip=True, fill_mode='nearest'
)

train_dir = os.path.join(base_dir, 'train')
target_count = 500 # العدد المطلوب لكل مرض

for cls in classes:
    cls_path = os.path.join(train_dir, cls)
    imgs = [f for f in os.listdir(cls_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    if len(imgs) < target_count:
        needed = target_count - len(imgs)
        print(f"جاري زيادة صور مرض {cls}...")
        for i in range(needed):
            img_name = np.random.choice(imgs)
            img = load_img(os.path.join(cls_path, img_name))
            x = img_to_array(img).reshape((1,) + (img.size[1], img.size[0], 3))

            # حفظ الصور الجديدة
            for batch in datagen.flow(x, batch_size=1, save_to_dir=cls_path, save_prefix='aug', save_format='jpg'):
                break
print("✅ اكتملت عملية التوازن!")
```

```python
import os
import random

# المسار الخاص بفولدر التدريب
train_dir = '/content/drive/MyDrive/Dental_Final_Split/train'
target_count = 500 # هنثبت كله على 500

print("جاري موازنة الفئات في فولدر التدريب...")

for cls in os.listdir(train_dir):
    cls_path = os.path.join(train_dir, cls)
    if os.path.isdir(cls_path):
        # جلب قائمة الصور
        images = [f for f in os.listdir(cls_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        current_count = len(images)

        if current_count > target_count:
            print(f"🔄 تقليل صور {cls}: من {current_count} إلى {target_count}")
            random.shuffle(images)
            images_to_remove = images[target_count:]
            for img in images_to_remove:
                os.remove(os.path.join(cls_path, img))
        elif current_count < target_count:
            print(f"⚠️ تنبيه: {cls} يحتوي على {current_count} صورة فقط (أقل من المستهدف)")
        else:
            print(f"✅ {cls} متوازن بالفعل (500 صورة)")

print("\n✨ انتهت عملية الموازنة بنجاح! كل الفئات الآن جاهزة.")
```

```python
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

# 1. إعداد الـ Data Generators
train_dir = '/content/drive/MyDrive/Dental_Final_Split/train'
val_dir = '/content/drive/MyDrive/Dental_Final_Split/val'

# التدريب بيحتاج Augmentation بسيط
train_datagen = ImageDataGenerator(rescale=1./255, rotation_range=20, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(train_dir, target_size=(224, 224), batch_size=32, class_mode='categorical')
val_gen = val_datagen.flow_from_directory(val_dir, target_size=(224, 224), batch_size=32, class_mode='categorical')

# 2. بناء الموديل
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(8, activation='softmax')(x) # عندك 8 أمراض/حالات
model = Model(inputs=base_model.input, outputs=predictions)

# تجميد الطبقات الأساسية
for layer in base_model.layers: layer.trainable = False

model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# 3. التدريب
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    brightness_range=[0.5, 1.5],
    shear_range=0.2,
    channel_shift_range=30.0,
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/Dental_Final_Split/train',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

val_gen = val_datagen.flow_from_directory(
    '/content/drive/MyDrive/Dental_Final_Split/val',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```
